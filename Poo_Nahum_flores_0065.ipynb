### conceptos de POO (Progamacion Orientada a Objetos) en phyton

##el prompt de gemini Ia
lenguaje Python  crear la clase "animal" con los atributos "nombre" ,"raza"crear el contructor  animal y las funciones ladrar(),correr(), crear los objetos pitbull, chihuahua, nivel principiante , no es interactivo, proporciona tu los datos

```python
class Animal:
    # El Constructor
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función (Método) Ladrar
    def ladrar(self):
        print(f"{self.nombre} está ladrando: ¡Guau, guau!🐻")

    # Función (Método) Correr
    def correr(self):
        print(f"{self.nombre} de raza {self.raza} está corriendo velozmente🏃‍♂️")

    def Morder(self):
        print(f"{self.nombre} está mordiendo a la vecina👩.")

# --- Creación de Objetos con datos proporcionados ---

# Objeto 1: Pitbull
pitbull = Animal("Rocky", "Pitbull")

# Objeto 2: Chihuahua
chihuahua = Animal("Chispa", "Chihuahua")
# --- objeto: perro callejero
perrocallejero = Animal("firulais", "callejero")

# --- Uso de los objetos ---

# Acciones del Pitbull
pitbull.ladrar()
pitbull.correr()
pitbull .Morder()
print("-" * 20) # Separador visual

# Acciones del Chihuahua
chihuahua.ladrar()
chihuahua.correr()
chihuahua.Morder()
print("-" * 20) # Separador visual

# Acciones del perro callejero
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.Morder()
print("progama realizado por Nahum flores NL = 18")
```
